Что такое **машинное обучение**?

Давайте подойдем к этому вопросу со стороны анализа названия:

Думаю, эпитет "машинное" всем понятен - имеется в виду, что человечество постоянно сталкивается с рутинными и монотонными задачами, которые не выходит автоматизировать классическими методами. Такие задачи отнимают много времени и сил у тех, кто решает их, при том они совершенно не приносят им удовольствия и морального удовлетворения от проделанной работы. Люди хотят научиться автоматизировать такую работу, то есть делегировать ее машине.

Что же означает слово "*обучение*"?

Как обычно мы понимаем это слово?
Скажем, когда ребенок учится алфавиту, в чем заключается его обучение?
Он должен запомнить, какой букве какой звук соответствует, а затем начать воспроизводить это соответствие. Как каждый из нас когда-то смог добиться этого?
*При помощи метода проб и ошибок!* Каждый раз, когда ребенок угадывает соответствие между буквой и звуком, он немедленно получает подтверждение этому от взрослых напрямую, либо же наблюдая за тем, как они читают ему вслух. Каждый же раз, когда он ошибается, он получает соответствующее опровержение, и ему приходится переосмыслить свой вывод.

То есть **обучение** в этом примере состоит в последовательно выстроенном ряде предположений, в попытке найти некую скрытую и неизвестную заранее зависимость. Выстраивание же происходит за счёт неких *штрафов* и *поощрений* со стороны окружающей среды.

Именно этого мы и хотим добиться от автоматизированной системы: также как ребенок, обучающийся чтению, получающий от взрослых подтверждения и опровержения своих гипотез, наша система должна искать сложные и неизвестные зависимости в данных, получая подтверждения и опровержения собственных гипотез.

## Машинное обучение с учителем

В этой постановке в качестве обучающей выборки нам дается совокупность некоторых объектов, каждый из которых мы можем каким-то образом численно описать. К каждому из этих объектов *задан некоторый вопрос*, а в обучающей выборке известны ответы на этот вопрос. Цель алгоритма состоит в том, чтобы по обучающей выборке понять, каким образом по описанию объекта можно дать ответ на заданный заранее вопрос.

Если говорить формально,

Обучающая выборка $D = \{(x_i, y_i)\}$ состоит из пар объект $x_i$ - метка $y_i$ и существует некий **закон природы**, 
который связывает $x$ и $y$: $f(x_i) = y_i$

Мы этот закон природы не знаем, но хотим, чтобы наш алгоритм *догадался*, что же это за закон, то есть хотим построить такой алгоритм $A$, чтобы $A(x) = f(x)$ для всех объектов $x$, какие мы только можем найти в природе.

Сложно?

Что такое признаковое описание объекта, тренировочная выборка, закон природы и алгоритм в нашем случае?

Признаковое описание каждой картины - это ответы на 3 поставленных вопроса. Мы задаем их для каждой картины и получаем верные ответы. То есть $x_i = (да/нет, да/нет, да/нет)$

Метка же, которую мы пытаемся по этим вопросам определить - это фамилия художника. То есть $y_i \in \{Айвазовский, Малевич, Брюллов\}$

Наш алгоритм - это искусствовед. Он всю жизнь учился определять картины, подмечал их особенности, и умеет по заданным вопросам определять художника.

Закон природы - это **настоящее соответствие** картин в упаковках и их авторов, которое пытался угадать искусствовед.

Обучающей выборкой же были все картины, которые он видел в жизни:
каждую из них он внимательно рассматривал, подмечал нужные детали, а затем имел возможность подойти к картине и достоверно узнать, кто был автором картины. То есть на этапе **обучения** алгоритм имел доступ к правильным ответам, а на этапе работы - не имел.

# Какие бывают признаки?
Мы постоянно говорим что-то об объектах, которые входят в наши наборы данных, но что вообще такое объект?

Понятно, что объект изначально представляет собой какую-то целостную сущность, но не можем же мы попросить алгоритм обработать объект произвольной природы. Не можем же мы предоставить компьютеру для анализа гербарий цветов или целый городской квартал. Для того чтобы научить компьютер работать с объектами произвольной природы, мы должны эти объекты формализовать, то есть составить их **признаковое описание**.

Признаковое описание объекта - это набор сведений об объекте, представленных в виде некоторого *вектора* (в данном случае -  множества чисел) значений.

Каждую координату этого вектора будем называть *признаком*. Каждый признак описывает какую-то характеристику объекта.

**Пример:**

Пусть наш объект - это автомобиль.

Какие мы можем придумать признаки для его описания?

Например, это может быть информация о цвете автомобиля, о мощности его двигателя, о наличии аудиосистемы, о максимальной скорости, массе и, скажем, марке.

Тогда конкретная совокупность таких признаков:
- Красный
- 100 л/с
- аудиосистема есть
- 150 км/ч
- 800кг
- Mercedes


задаст некоторый *конкретный автомобиль*.

То есть мы как бы подменяем для себя объект некоторой совокупностью информации, которую мы о нем собрали, и можем записать в виде последовательности чисел.

Вы спросите: как же так, ведь "зеленый", "аудиосистема есть", "Mercedes" - все это не числа. Как тут быть?

Прибегнем к следующей хитрости:

Пусть мы знаем, что автомобили, которые нас интересуют могут быть лишь конкретных марок и конкретных цветов (марок в мире конечное число, как и цветов краски для автомобиля, так что это множество точно не будет бесконечным).

Например, цвета могут быть - зеленый, красный и синий, а марки - mercedes, peugeot и lada.

Пронумеруем эти цвета и марки:

0.Зеленый      0.Mercedes

1.Красный      1.Peugeot

2.Синий        2.Lada

Теперь мы можем смело подменить признак "Красный" на значение 1, а признак "Mercedes" на значение 0. Вместо "аудиосистема есть" будем писать 1, а в случае, когда ее нет - будем писать 0

Теперь наш автомобиль задан следующим вектором, который полностью его описывает:

$[1, 100, 1, 150, 800, 0]$

Такой вектор уже можно отправлять на обработку численным алгоритмам.

Для обучения модели на обучающей выборке $X$ с вектором правильных ответов $y$ используется вызов метода `.fit(X,y)` этого класса. Для получения предсказаний алгоритма мы пользуемся методом `.predict(X)`, а если хотим получить распределение вероятностей отнести ответ к тому или иному классу, мы вызываем `.predict_proba(X)`

Итак, наш алгоритм следующий:

* Первым делом мы загрузим датасет при помощи функции `load_iris`. Мы внимательнно присмотримся к этому датасету, визуализируем его;
* Далее мы разобьём датасет на тренировочную и тестовую часть при помощи функции `train_test_split`. Зачем нужно такое преобразование, мы подробно разберём на одной из следующих лекций;
* Затем мы обучим алгоритм на обучающей выборке и протестируем его качество на тестовой выборке.

Начнём!

```python
# Импорт необходимых модулей из библиотеки sklearn, а также pandas и seaborn
import sklearn
import warnings
import pandas as pd
import seaborn as sns
sns.set_theme()
warnings.filterwarnings("ignore")

from sklearn.datasets import load_iris
from sklearn.metrics  import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Загрузка датасета
X, y = load_iris(return_X_y=True)
```

Визуализируем датасет при помощи инструмента `sns.pairplot`. В качестве визуализации выберем инструмент, позволяющий изобразить все точки датасета в координатах, отвечающих всевозможным парам признаков. Подробнее с этим инструментом можно ознакомиться по [ссылке](https://seaborn.pydata.org/generated/seaborn.pairplot.html).

```python
"""
При описании цветка используются следующиее признаки:

1. Длина наружной доли околоцветника (англ. sepal length);
2. Ширина наружной доли околоцветника (англ. sepal width);
3. Длина внутренней доли околоцветника (англ. petal length);
4. Ширина внутренней доли околоцветника (англ. petal width).
"""
iris = pd.DataFrame({'sepal length' : X[:,0],
                     'sepal width' : X[:,1],
                     'petal length' : X[:,2],
                     'petal width' : X[:,3],
                     'target' : y})

sns.pairplot(data=iris, hue='target')
```

Сможем ли мы отличить цветы разных видов по такому признаковому описанию?
Попробуем сделать это, применив алгоритм логистической регрессии.

```python
# Разбиение выборки на тренировочную и тестовую
X_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, shuffle=True)

# Обучение модели
model = LogisticRegression().fit(X_train, y_train)
```

```python
model
```

Сделаем предсказания относительно тестовой выборки

```python
preds = model.predict(x_test)
preds
```

```python
y_test
```

Нам необходимо проверить, насколько наша модель оказалась качественной. Для этого мы воспользуемся функцией `accuracy_score`, вычисляющей долю верных ответов алгоритма.

```python
acc = accuracy_score(y_test, preds)
acc
```

Заметьте, в предыдущем примере про автомобили признаки существенно отличались друг от друга. Какими же они бывают вообще?

На самом деле, типов признаков довольно много. Почему важно отличать их друг от друга? Потому что для работы с каждым из них применяются свои собственные методы, не похожие друг на друга.



### 1. Бинарные признаки

Бинарные признаки - это признаки, которые могут принимать лишь 2 значения. Пример такого признака - это наличие аудиосистемы в автомобиле. Да и вообще, они обычно используются, как метка *наличия* чего-либо у исследуемого объекта. Помимо этого, можно помечать бинарными признаками некоторые полярные свойства объекта - скажем, биологический пол (мужчина и женщина), северное или южное полушарие, положительный или отрицательный заряд частицы.

Такие признаки обычно помечают комбинацией значений 0 и 1, либо -1 и 1



### 2. Категориальные признаки

Категориальные признаки немного сложнее, чем бинарные. Отличие состоит в том, что категориальные признаки могут принимать значения из некоторого конечного множества. Как цвет или марка автомобиля. Тема новостной статьи или оценка школьника также представляют собой категориальный признак.

Однако, категориальные признаки и сами делятся на 2 класса: *номинальные* и *ранговые*.

В случае номинальных признаков, не имеет места сравнение их друг с другом, за исключением сравнения на равенство. То есть фраза "синий цвет больше, чем красный" не имеет для нас никакого смысла. Мы не может сравнить, какой цвет больше, а какой меньше.

А вот фраза "майор больше (старше) сержанта" - уже имеет. Воинские звания ведь такой же категориальный признак, однако их мы уже можем сравнивать друг с другом, и это будет осмысленно. Поэтому такой признак называется ранговым.

Категориальные признаки могут быть закодированы при помощи нумерации исходного признакового пространства.



### 3. Непрерывные

Это самый понятный тип признаков. Мы что-то померили, взвесили, посчитали и получили некоторое число, которое каким-то образом описывает наш объект. Это число будет принадлежать некоторому диапазону, то есть будет взято из непрерывного подмножества множества действительных чисел.

Примером является масса автомобиля (она может быть как ровно 800кг, так и 799.54 кг, 802.345 кг - и вообще никаких ограничений).



### 4. Географические данные

Это данные формата широта-долгота на поверхности Земли.

Конечно, такие признаки одновременно и довольно редки, и достаточно сложны в обработке в связи со своей специфичностью. Они обычно возникают в задачах, непосредственно связанных с логистикой. Мы будем рассматривать пример такой задача в одной из следующих лекций.

# Что такое модель машинного обучения?

Каждый слышал про модели, которые специалисты ML старательно обучают. Что же это за звери такие, которых можно чему-то научить?

Мы говорили, что наша итоговая цель - это придумать такой алгоритм $A(x)$, который будет очень сильно похож на скрытый от наших глаз закон природы. А на что должен быть похож этот алгоритм? Это мы решаем сами. Выбор тех функций, из которых мы будем брать наш алгоритм, и называется выбором модели.

Модель - это *семейство* функций, зависящих от некоторых параметров, из которых мы будем выбирать наш алгоритм.

**Пример:**
Пусть настоящая неизвестная зависимость имеет вид, похожий на параболу. Мы будем выбирать алгоритм регрессии из множества многочленов второй степени $$A(x) = ax^2 + bx + c$$
Это и будет наша модель.

Когда мы выберем наилучшую из всех возможных функций из данного семейства, она и будет нашей решающей функцией или алгоритмом.

# Постановка задачи машинного обучения

Теперь мы можем формально поставить задачу машинного обучения и обсудить, на каких идеях основано решение.


## Формальная постановка задачи


**Задача машинного обучения** - это задача построения алгоритма решения некоторой проблемы по известному набору прецедентов. То есть на основе известного накопленного опыта мы должны научиться принимать некоторые решения в ситуациях, которые мы никогда не встречали.

В общем случае формализация данной задачи выглядит так:

1) Зададимся некоторой функцией $L_a(x, y)$, которая позволяет по правильному ответу $y$, отнесённому к объекту $x$, вычислить значение **ошибки** алгоритма $a$ при обработке этого объекта. Такую функцию называют **функцией потерь**, или **лосс-функцией**

Пример:

Пусть мы решаем задачу регрессии, по объекту $x$ необходимо восстановить некоторое значение $y$ (например, $x$ - это признаковое описание автомобиля, а $y$ - значение его стоимости).

Пусть есть некотрый алгоритм $a(x)$, позволяющий предсказывать ожидаемое значение $y$ по $x$.

Тогда примером функции потерь может служить модуль отклонения предсказанного значения от истинного: $$L_a(x,y) = |a(x) - y|$$



## Сведение к задаче оптимизации

2) Пусть мы умеем измерять ошибку алгоритма $a$ при обработке произвольного объекта $x$ при условии, что мы знаем верный ответ $y$ при помощи введённой в п.1 функции потерь. Какова наша цель?

Очевидно, мы хотим выбрать такой алгоритм, который будет меньше всего ошибаться, причем не только на данной нам тренировочной выборке, но и **вообще на любых предложенных ему для обработки объектах**. Однако, не имея информации о всех возможных объектах $x$, мы ограничиваемся вычислением ожидаемой ошибки на известном нам множестве, то есть на нашей тренировочной выборке.

3) В качестве ожидания ошибки на тренировочной выборке мы вычислим среднее значение функции потерь на ней. То есть наша задача сводится к минимизации среднего значения функции потерь $$\frac{1}{N}∑\limits_{x \in X}L_a(x, y_x) → \min\limits_a$$
где $N$ - это количество объектов в тренировочной выборке $X$.

Получается, что мы свели задачу машинного обучения к задаче оптимизации (минимизации) средней функции потерь. Решение этой задачи лежит в области теории оптимизации. Мы рассмотрим некоторые методы теории оптимизации в рамках следующих лекций.




## Дополнительно (для мат.факультетов)*:
___________________________________________
Говоря формально, мы хотим минимизировать *математическое ожидание функции потерь на пространстве всевозможных объектов*. Для слушателей, знакомых с классической теорией вероятностей и теорией меры, отметим, что записать эту постановку можно следующим образом:

$$R(a) = ∫\limits_{X}L_a(x,y)dP(x) → \min\limits_{a}$$

В данном случае $P(x)$ - это вероятностная мера, введённая на пространстве объектов $X$, отображающая факт того, что некоторые из объектов могут попадаться в реальности чаще, а некоторые - реже.

Выражение $R(a)$ называется **ожидаемым риском, ассоциированным с алгоритмом $a$**. Конечно, посчитать значение $R(a)$ не представляется возможным, но есть способы оценить это значение. В книге Вапника В.Н. и Червоненкиса А.Я. "Теория распознавания образов (статистические проблемы обучения)" представлены исторически сформировавшиеся пути оценки ожидаемого риска.

Мы же остановимся на самом популярном случае оценки $R(a)$ через тренировочную выборку.

**Эмпирическим риском $Q(a,X)$ называется оценка $R(a)$ при помощи вычисления выборочного мат.ожидания на тренировочной выборке:**
$$Q(a, X) = \frac{1}{N}∑\limits_{x \in X}L_a(x, y_x)$$

где $N$ - это количество объектов в выборке, а под $y_x$ подразумевается правильный ответ, отнесённый к объекту $x$.

Принцип **минимизации эмпирического риска** - основной принцип решения задачи обучения по прецедентам. То есть задача машинного обучения заменяется в такой формулеровке на задачу оптимизации:
$$Q(a, X) → \min\limits_a$$

Замечание 1.

Мы предполагаем, что алгоритм $a$ мы ищем не вполне произвольный, а относящийся к некоторому многопараметрическому семейству функций, то есть рассматриваем конкретную **модель машинного обучения $A$**. Пусть это семейство функций параметризуется набором параметров $Θ$, тогда $$Θ^* = arg \min\limits_{Θ}Q(a_{Θ}, X)$$
где $Θ^*$ - оптимальные параметры, а $a_{Θ} \in A$ - алгоритм из выбранного семейства $A$, параметризованный набором параметров $Θ$

Замечание 2.

В рамках метода минимизации эмпирического риска много усилий приходится применять для анализа и оценки репрезентативности тренировочной выборки, поскольку на шаге перехода от минимизации $R(a)$ к минимизации $Q(a,X)$ по сути делается предположение о том, что тренировочная выборка "хорошая".

* $N$ - мы обычно будем обозначать число строк в датасете (или, что то же самое, количество объектов в соответствующей выборке);
* $M$ - количество столбцов то есть число (признаков);
* $a(x)$ обычно будет обозначать применение алгоритма машинного обучения $a$ к объекту $x$;
* $x$ (малый) обычно обозначает некоторый объект или его признаковое описание (то есть вектор, его описывающий);
* $X$ (большой) представляет из себя набор объектов или матрицу их признаковых описаний;
* $y$ - метка для соответствующего объекта или вектор меток для набора объектов;

* $L_a(x, y)$ - значение функции потерь алгоритма $a$ на объекте $x$, которое отвечает вектору правильных ответов $y$. Обычно мы не будем писать так сложно: из контекста почти всегда ясно, о каком алгоритме идёт речь, а также само собой разумеется, что $y$ - это ответы из имеющейся у нас выборки. Поэтому чаще мы будем писать $L(x)$, опуская лишние обозначения, а в случаях, когда понятно, о каком конкретно объекте $x$ идёт речь, мы можем вообще ограничиться обозначением $L$;

* $Q(x)$ - ожидаемое значение функции потерь на объекте $x$. Проще говоря, $Q(x)$ - это оценка ошибки при обработке объекта $x$.
